```python
numero1 = int(input("Escolha um número:"))
print(numero1)
```

```python
def somar_numero(numero1):
  numero2 = int(input("Escolha o número para somar com o primeiro número"))
  numero1 += numero2
  print(numero1)

def diminiur_numero(numero1):
  numero2 = int(input("Escolha o número para súbtrair com o primeiro número"))
  numero1 -= numero2
  print(numero1)

def multiplicar_numero(numero1):
  numero2 = int(input("Escolha o número para multiplicar com o primeiro número"))
  numero1 *= numero2
  print(numero1)

def dividir_numero(numero1):
  numero2 = int(input("Escolha o número para dividir com o primeiro número"))
  numero1 /= numero2
  print(numero1)

def fatorar_numero(numero1):
  fatorial = 1
  for i in range(2, numero1 + 1):
      fatorial *= i
  print(fatorial)

def raiz_numero(numero1):
  import math
  resultado = math.sqrt(numero1)
  print(resultado)
```

```python
print(numero1)
print("A - Somar números")
print("B - Súbtrair números")
print("C - Multiplicar números")
print("D - Dividir números")
print("E - Fatorar número")
print("F - Raiz do número")
opcao = input("Escolha uma opção:")
```
